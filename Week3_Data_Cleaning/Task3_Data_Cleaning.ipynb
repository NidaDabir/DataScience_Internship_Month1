```python
import pandas as pd
import numpy as np

df = pd.read_excel("C:/Users/Nida/Downloads/Sample-Superstore dataset.xlsx")

df.head()
```

# We will insert missing values into some columns.

```python
# Insert missing values randomly in Sales column
df.loc[df.sample(frac=0.05).index, 'Sales'] = np.nan

# Insert missing values in Profit column
df.loc[df.sample(frac=0.05).index, 'Profit'] = np.nan

# Insert missing values in Category column
df.loc[df.sample(frac=0.03).index, 'Category'] = np.nan

df.isnull().sum()
```

```python
# handling Missing values

# 1.Using Mean (for Sales)
df['Sales'].fillna(df['Sales'].mean(), inplace=True)
```

```python
# 2. Using Median (for Profit)
df['Profit'].fillna(df['Profit'].median(), inplace=True)
```

```python
# 3. Using Mode (for Category)
df['Category'].fillna(df['Category'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

# Removing Duplicate Rows

```python
# First we will intentionally create duplicates:

df = pd.concat([df, df.iloc[:5]], ignore_index=True)
df.duplicated().sum()
```

```python
# Now remove duplicates:

df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

# Converting Incorrect Data Type

```python
# Let’s intentionally convert Sales to string first:
df['Sales'] = df['Sales'].astype(str)
df.dtypes
```

```python
# Now convert back to numeric:
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df.dtypes
```

```python
df.to_csv("Cleaned_Superstore_Dataset.csv", index=False)
```
