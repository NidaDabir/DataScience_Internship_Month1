```python
# Profit Margin (Calculated Column 1)
df["Profit_Margin"] = df["Profit"] / df["Sales"]
```

```python
# Sales Per Quantity (Calculated Column 2)
df["Sales_Per_Unit"] = df["Sales"] / df["Quantity"]
```

```python
# Discounted Sales (Calculated Column 3)
df["Discounted_Sales"] = df["Sales"] * (1 - df["Discount"])
```

```python
df.head()
```

```python
# Filtering Rows Based on 2 Conditions.
# Example: Region = West, Profit > 100

filtered_df = df[(df["Region"] == "West") & (df["Profit"] > 100)]
filtered_df.head()
```

```python
# Group By + Mean + Count
# Group by Category:

grouped = df.groupby("Category").agg({
    "Sales": "mean",
    "Profit": "mean",
    "Order ID": "count"
})

grouped
```

```python
# Renaming count column:
grouped.rename(columns={"Order ID": "Total_Orders"}, inplace=True)
grouped
```
