```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Nida\PyCharmMiscProject\Cleaned_Superstore_Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

# PLOT 1 — Distribution of Sales

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

# PLOT 2 — Sales by Category (Bar Plot)

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.xticks(rotation=45)
plt.show()
```

# PLOT 3 — Profit by Region

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Region', y='Profit', data=df)
plt.title("Profit Distribution by Region")
plt.show()
```

# PLOT 4 — Sales vs Profit (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()
```

# PLOT 5 — Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Insight 1
Most sales values are concentrated in lower ranges, with few high-value outliers.

# Insight 2
Technology category generates higher average sales compared to Furniture and Office Supplies.

# Insight 3
Certain regions show higher profit variability, indicating inconsistent performance.

# Insight 4
There is a positive correlation between Sales and Profit — higher sales often lead to higher profit.

# Insight 5
Discount has a weak or negative correlation with Profit, meaning heavy discounts may reduce profitability.
